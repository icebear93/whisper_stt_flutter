# 파이널 컷에서 mp3 파일 추출까지 완료 한 상태에서 아래 시작

```python
import whisper
import warnings
warnings.filterwarnings('ignore')
import datetime
import srt
```

```python
## 테스트 출력

model = whisper.load_model('large-v2')
result = model.transcribe('New Recording 27.m4a', language = 'ko')
print(result['text'])
```

# Main

```python
def seconds_to_srt_time_format(seconds):
    return datetime.timedelta(seconds=seconds)

def generate_srt_from_transcription(transcription):
    subtitles = []
    for i, segment in enumerate(transcription['segments']):
        start_time = seconds_to_srt_time_format(segment['start'])
        end_time = seconds_to_srt_time_format(segment['end'])
        text = segment['text']

        subtitle = srt.Subtitle(index=i+1, 
                                start=start_time,
                                end=end_time,
                                content=text)
        subtitles.append(subtitle)

    return srt.compose(subtitles)

# Whisper 모델 로드
model = whisper.load_model('large-v2')

# 음성 파일에서 자막 추출
result = model.transcribe('/Users/jsp/Desktop/week4.mp3', language='ko', verbose=False) # ==========================>>>>>>  stt 할 파일 넣기

# 자막과 타임코드를 srt 형식으로 변환
srt_content = generate_srt_from_transcription(result)

# srt 파일로 저장
with open('output_subtitles.srt', 'w') as f: # ==========================>>>>>>  srt 이름 설정
    f.write(srt_content)

print("SRT 파일이 생성되었습니다.")
```
